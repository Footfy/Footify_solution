```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Carregar os dados do arquivo CSV
df = pd.read_csv('produtos.csv')

# 2. Encontrar o menor preço para cada produto e a loja correspondente
menores_precos = df.loc[df.groupby('nome')['preco'].idxmin()]

print("\nMenor preço por produto com a respectiva loja:")
print(menores_precos)

# 3. Pivotar os dados para visualizar os preços de cada loja por produto
pivot_df = df.pivot_table(index='nome', columns='loja', values='preco')

# 4. Visualizar os preços com gráfico de barras
pivot_df.plot(kind='bar', figsize=(12, 6))
plt.title('Comparação de Preços entre Lojas por Produto')
plt.ylabel('Preço (R$)')
plt.xlabel('Produto')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend(title='Loja')
plt.tight_layout()
plt.show()

# 5. Salvar os menores preços em um novo CSV
menores_precos.to_csv('menores_precos_por_produto.csv', index=False)
print("\nArquivo 'menores_precos_por_produto.csv' gerado com sucesso!")
```
